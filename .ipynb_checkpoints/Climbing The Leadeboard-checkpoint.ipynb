```python
#import copy
#import time
# Upload the code as a file in HackerRank
# Uncoment bewlow for Custom input
scores_count = int(input())

scores = list(map(int, input().rstrip().split()))

alice_count = int(input())

alice = list(map(int, input().rstrip().split()))

start=time.time()
def arrangeScoresRank(scores):
    rank_arrange=[scores[0]]
    for i in range(1,len(scores)):
        if(scores[i-1]!=scores[i]):
            rank_arrange.append(int(scores[i]))
    return rank_arrange


scores_singles=arrangeScoresRank(scores)
l=len(scores_singles)
l_const=l

def climbingLeaderboard(scores_singles, alice):
    ranks_of_alice=[]
    l=len(scores_singles)
    l_const=(l)
    for i in alice:
        if(i==scores_singles[l_const-1]):
            ranks_of_alice.append(l_const)
            continue
        elif(i<scores_singles[l_const-1]):
            ranks_of_alice.append(l_const+1)
            continue
        elif (i>scores_singles[0]):
            ranks_of_alice.append(1)
            continue
        elif (i==scores_singles[0]):
            ranks_of_alice.append(1)
            continue
        for j in range(l-1,-1,-1):
            
            if (i>=scores_singles[j]):
                continue
            else:
                ranks_of_alice.append(j+1+1)
                l=j+2
                break

    return ranks_of_alice
print('\n'.join(map(str, climbingLeaderboard(scores_singles,alice))))
```
